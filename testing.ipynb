```python
def print_layers(model, prefix=""):
    layer_names = []
    for name, layer in model.named_children():
        new_prefix = f"{prefix}.{name}" if prefix else name
        layer_names.append(new_prefix)
        layer_names += print_layers(layer, prefix=new_prefix)
    return layer_names
```
